```python

'''Write a Python program to calculate the gross salary of an employee
If basic <= 10000: HRA = 20% , DA = 80%
If basic <= 20000: HRA = 25% , DA = 90%
otherwis: HBA = 30% , DA = 95%'''

basic = float(input('Enter basic salary: '))

if basic <= 10000:
    hra = basic * 0.2
    da = basic * 0.8
elif basic <= 20000:
    hra = basic * 0.25
    da = basic * 0.9
else:
    hra = basic * 0.3
    da = basic * 0.95

gross_salary = basic + hra + da

print(f'Basic Salary: {basic}')
print(f'HRA: {hra}')
print(f'DA: {da}')
print(f'Gross Salary: {gross_salary}')
```

```python
''' Write a Python program to calculate income tax based on annual income:
Up to rupees 2,50,000: No tax
rupees 2,50,001- rupees 5,00,000: 5%
rupees 5,00,001- rupees 10,00,000: 20%
Above rupees 10,00,000: 30%'''

income = float(input("Enter annual income: "))

if income <= 250000:
    tax = 0
elif income <= 500000:
    tax = (income-2,50,000)*0.5
elif income <= 1000000:
    tax = (250000 * 0.05) + (income - 500000)*0.20
else:
    tax = (250000 * 0.05) + (500000 * 0.20) + (income - 1000000)*0.30

print(f"Annual Income: {income}")
print(f"Income Tax: {tax}")
```
